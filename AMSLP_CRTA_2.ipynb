```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
```

```python
# Preprocesar el texto
# Preprocesar el texto
def preprocess(text):
    if isinstance(text, str):  # Check if the input is a string
        doc = nlp(text)
        return " ".join([token.lemma_ for token in doc if not token.is_stop and not token.is_punct])
    else:
        return ""  # Return an empty string for non-string inputs

df_san_luis['processed'] = df_san_luis['Contenido'].apply(preprocess)
```

```python
# Vectorización con TF-IDF usando n-gramas de caracteres de 3 a 5 longitudes
vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(3, 5))
X = vectorizer.fit_transform(df_san_luis['processed'])

# Normalización de vectores
X_normalized = normalize(X)
```

```python
# Clustering
db = DBSCAN(eps=0.5, min_samples=2, metric='cosine')
clusters = db.fit_predict(X_normalized)
```

```python
# Clustering con K-means
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X_normalized)
```

```python
df_san_luis['cluster'] = clusters

# Análisis de clusters
cluster_counts = df_san_luis['cluster'].value_counts()

# Imprimir los resultados de los clusters más grandes
print(cluster_counts)
```

```python
cluster_texts = {}
for i in range(min(len(clusters), len(df_san_luis))): # Ensure you don't exceed the DataFrame length
  if clusters[i] not in cluster_texts:
    cluster_texts[clusters[i]] = []
  cluster_texts[clusters[i]].append(df_san_luis['Contenido'].iloc[i]) # Use .iloc for integer-based indexing

cluster_dfs = {}
for cluster_id, texts in cluster_texts.items():
  cluster_dfs[cluster_id] = pd.DataFrame({'Contenido': texts})
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
def generar_nube_de_palabras(textos_limpios):

    # Unir todos los cuerpos de texto en una sola cadena
    textos_unidos = ' '.join(textos_limpios)

    # Crear la nube de palabras
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(textos_unidos)

    # Mostrar la nube de palabras
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
generar_nube_de_palabras(df_cluster_3['clean_body'])
```
